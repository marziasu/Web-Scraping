```python
import pandas as pd
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import time
```

```python
s=Service('C:/Users/Marzia/Documents/web scrape/chromedriver.exe')
driver= webdriver.Chrome(service=s)
```

```python
df=pd.DataFrame()
for j in range(1,101):
    url=f'https://www.linkedin.com/search/results/companies/?companyHqGeo=[%22106215326%22]&keywords=software%20development%20in%20bangladesh&origin=FACETED_SEARCH&page={j}&searchId=263dc6a5-5b10-45e2-9e4e-e6ed1493ed91&sid=Sri'
    driver.get(url)
    html_content = driver.execute_script('return document.body.innerHTML')
    soup = BeautifulSoup(html_content, "html.parser")
    company= soup.find_all('li', class_='reusable-search__result-container')
    name=[]
    category=[]
    followers=[]
    description=[]
    for i in company:
        name.append(i.find_all('div', class_='display-flex')[1].text.strip() )
        category.append(i.find('div', class_='entity-result__primary-subtitle').text.strip() )
        followers.append(i.find('div', class_='entity-result__secondary-subtitle').text.strip() )
        description.append(i.find('p', class_='entity-result__summary').text.strip() )
    df1= pd.DataFrame(
        {
            "Name": name,
            "Category": category,
            "Followers": followers,
            "Description": description
        })
    df=pd.concat([df,df1])
    print( df.shape)
    time.sleep(4)

```
